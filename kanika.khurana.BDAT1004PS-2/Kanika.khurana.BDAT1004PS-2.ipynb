```python
def fileLength(filename):
    try:
        infile = open(filename, 'r')
        content = infile.read()
        infile.close()
        print(len(content))
    except FileNotFoundError:
        print(f"File {filename} not found.")

        
fileLength('ana.ipynb')  # Existing file
fileLength('filelength.txt')  # Existing file
fileLength('length.txt')  # NON-Existing file
```

```python
class Marsupial:
    def __init__(self):
        self.pouch_contents = []

    def put_in_pouch(self, item):
        self.pouch_contents.append(item)

    def get_pouch_contents(self):
        return self.pouch_contents


class Kangaroo(Marsupial):
    def __init__(self, x, y):
        super().__init__()
        self.x = x
        self.y = y

    def jump(self, dx, dy):
        self.x += dx
        self.y += dy

    def __str__(self):
        return f"I am a Kangaroo located at coordinates ({self.x},{self.y})"


m = Marsupial()
m.put_in_pouch('doll')
m.put_in_pouch('firetruck')
m.put_in_pouch('kitten')
print(m.get_pouch_contents())

k = Kangaroo(0, 0)
print(k)
k.put_in_pouch('doll')
k.put_in_pouch('firetruck')
k.put_in_pouch('kitten')
print(k.get_pouch_contents())
k.jump(1, 0)
k.jump(1, 0)
k.jump(1, 0)
print(k)
```

```python
from html.parser import HTMLParser


class HeadingParser(HTMLParser):
    def __init__(self):
        super().__init__()
        self.headings = []
        self.current_heading = ""
        self.indentation = 0
        self.valid_heading_tags = ["h1", "h2", "h3", "h4", "h5", "h6"]

    def handle_starttag(self, tag, attrs):
        if tag in self.valid_heading_tags:
            self.indentation = self.valid_heading_tags.index(tag)

    def handle_data(self, data):
        if self.indentation >= 0:
            self.current_heading += " " * self.indentation + data.strip()

    def handle_endtag(self, tag):
        if tag in self.valid_heading_tags:
            self.headings.append(self.current_heading)
            self.current_heading = ""

    def print_headings(self):
        for heading in self.headings:
            print(heading)


with open("w3c.txt", "r") as file:
    content = file.read()

hp = HeadingParser()
hp.feed(content)
hp.print_headings()
```

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin

def webdir(url, depth, indent):
    # Fetch the web page
    try:
        response = requests.get(url)
    except requests.exceptions.RequestException as e:
        print("Error:", e)
        return
    soup = BeautifulSoup(response.content, 'html.parser')
    print(' ' * indent + soup.title.string)
    links = soup.find_all('a')

    for link in links:
        href = link.get('href')
        if href and not href.startswith('#'):
            full_url = urljoin(url, href)
            print(' ' * (indent + 4) + full_url)
    if depth > 0:
        for link in links:
            href = link.get('href')
            if href and href.startswith('http'):
                webdir(urljoin(url, href), depth - 1, indent + 4)
webdir('https://example.com', 2, 0)
```

```python
import sqlite3
conn = sqlite3.connect('weather.db')
cursor = conn.cursor()

# Query a) All the temperature data
cursor.execute("SELECT Temperature FROM weather")
temperature_data = cursor.fetchall()
print("Temperature Data:")
for row in temperature_data:
    print(row[0])

# Query b) All the cities, but without repetition
cursor.execute("SELECT DISTINCT City FROM weather")
cities = cursor.fetchall()
print("\nCities:")
for row in cities:
    print(row[0])

# Query c) All the records for India
cursor.execute("SELECT * FROM weather WHERE Country='India'")
india_records = cursor.fetchall()
print("\nRecords for India:")
for row in india_records:
    print(row)

# Query d) All the Fall records
cursor.execute("SELECT * FROM weather WHERE Season='Fall'")
fall_records = cursor.fetchall()
print("\nFall Records:")
for row in fall_records:
    print(row)

# Query e) The city, country, and season for which the average rainfall is between 200 and 400 millimeters
cursor.execute("SELECT City, Country, Season FROM weather GROUP BY City, Country, Season HAVING AVG(Rainfall) BETWEEN 200 AND 400")
average_rainfall_records = cursor.fetchall()
print("\nRecords with Average Rainfall between 200 and 400:")
for row in average_rainfall_records:
    print(row)

# Query f) The city and country for which the average Fall temperature is above 20 degrees, in increasing temperature order
cursor.execute("SELECT City, Country FROM weather WHERE Season='Fall' GROUP BY City, Country HAVING AVG(Temperature) > 20 ORDER BY AVG(Temperature) ASC")
average_fall_temperature_records = cursor.fetchall()
print("\nRecords with Average Fall Temperature above 20 degrees:")
for row in average_fall_temperature_records:
    print(row)

# Query g) The total annual rainfall for Cairo
cursor.execute("SELECT SUM(Rainfall) FROM weather WHERE City='Cairo'")
total_rainfall_cairo = cursor.fetchone()[0]
print("\nTotal Annual Rainfall for Cairo:", total_rainfall_cairo)

# Query h) The total rainfall for each season
cursor.execute("SELECT Season, SUM(Rainfall) FROM weather GROUP BY Season")
total_rainfall_season = cursor.fetchall()
print("\nTotal Rainfall for Each Season:")
for row in total_rainfall_season:
    print(row[0], "-", row[1])
conn.close()
```

```python
words = ['The', 'quick', 'brown', 'fox', 'jumps', 'over', 'the', 'lazy', 'dog']

# a) 
uppercase_words = [word.upper() for word in words]
print(uppercase_words)
print()

# b) 
lowercase_words = [word.lower() for word in words]
print(lowercase_words)
print()

# c) 
word_lengths = [len(word) for word in words]
print(word_lengths)
print()

# d) 
word_info = [[word.upper(), word.lower(), len(word)] for word in words]
print(word_info)
print()

# e) 
long_words = [word for word in words if len(word) >= 4]
print(long_words)
print()
```
